## Soru-1
### İllere göre belediye hizmetlerinden memnuniyet anketi verilerini alın:
2013 yılında TÜİK tarafından yapılan illere göre belediye hizmetlerinden memnuniyet anketi verilerini, ekte verilen "2013_Belediye Hizmetleri Anketi_1.xls", "2013_Belediye Hizmetleri Anketi_2.xls" ve "2013_Belediye Hizmetleri Anketi_3.xls" dosyalarından yükleyin ve herbiri için birer DataFrame oluşturun ve aşağıdaki işlemleri uygulayın:

```
1- Dosyaların ilk 3 satırını ve son 2 satırını atlayın. Hiyerarşik indekslemeye uygun olarak okuyun. (ipucu: 'header' parametresi)
2- Her bir dataframe'deki boş sütunları (bütün değerleri 'NaN' olan sütunları) silin
3- Ortak sütun üzerinden bu üç veri kümesini birleştirin ve ortak sütunu index olarak belirleyin. Ayrıca index ismini 'İL ADI' olarak belirleyin.
4- Sütun başlıklarından İngilizce kelimeleri silin
5- Memnuniyet oranı olarak '-' belirtilen hücrelerin değerini 'NaN' iledeğiştirin

```

````python
import pandas as pd
import re
import numpy as np

df1=pd.read_excel('2013_Belediye Hizmetleri Anketi_1.xlsx',skiprows=3,skipfooter=2,header=[0,1])

df2=pd.read_excel('2013_Belediye Hizmetleri Anketi_2.xlsx',skiprows=3,skipfooter=2,header=[0,1])

df3=pd.read_excel('2013_Belediye Hizmetleri Anketi_3.xlsx',skiprows=3,skipfooter=2,header=[0,1])
df3=df3.dropna(axis=1)
df1=df1.dropna(axis=1)
df2=df2.dropna(axis=1)

dx =[df1,df2,df3]
final = pd.concat(dx,axis=1)


final.columns = final.columns.rename("İL ADI",level=1)

final=final.rename(columns=lambda x: x[:x.find("\n")] )
final=final.replace('-', final.replace('-', np.nan))

final
````

## Soru-5
### Yerel seçimlerde partilerin il bazında aldıkları belediye meclisi oy sayılarını alın:
2014 yılı yerel seçimlerinde partilerin il bazında aldıkları belediye meclisi oy sayılarını, ekte verilen "2014_Il Bazinda Belediye Meclisi Oy Sayilari.xlsx" dosyasından yükleyin ve başka bir DataFrame daha oluşturun ve aşağıdaki işlemleri uygulayın:

```
1- 'İL ADI' sütununu index olarak belirleyin. 
2- Boş sütunları (bütün değerleri 'NaN' olan sütunları) silin
3- İkinci sütununun ismini 'BÜYÜKŞEHİR' olarak değiştirin, ve o sütunda içeriği 'BÜYÜKŞEHİR' olan hücreleri 'EVET' ile, diğer hücreleri 'HAYIR' ile değiştirin
```

````python
import pandas as pd
import numpy as np
import re



secimDf=pd.read_excel('2014_Il Bazinda Belediye Meclisi Oy Sayilari.xlsx')
secimDf=secimDf.dropna(how='all', axis=1)
secimDf = secimDf.rename(columns = {'Unnamed: 1' : 'BÜYÜKŞEHİR'})
secimDf.set_index('İL ADI',inplace=True)
secimDf['BÜYÜKŞEHİR'] = secimDf['BÜYÜKŞEHİR'].replace({'BÜYÜKŞEHİR' : 'EVET'}, regex=True)
secimDf['BÜYÜKŞEHİR'] = secimDf['BÜYÜKŞEHİR'].replace(np.nan, 'HAYIR', regex=True)
secimDf
````

````python
def gecersizOranBolge(x):
    soru18df=soru16df
    test18df=soru18df[soru18df['Bölge'] == x]
    length=test18df.index.size
    kullananSecmen=0
    gecerliOy=0
    for j in range(length):
        kullananSecmen+=test18df.iat[j,2]
        gecerliOy+=test18df.iat[j,26]
    tempGecersizOran = (1-(gecerliOy/kullananSecmen))*100
    gecersizOran=float("{0:.2f}".format(tempGecersizOran))
    return gecersizOran

def katilimOranBolge(y):
    soru18df=soru16df
    test20df=soru18df[soru18df['Bölge'] == y]
    lengths=test20df.index.size
    kullananSecmenler=0
    secmenSayisi=0
    for j in range(lengths):
        kullananSecmenler+=test20df.iat[j,2]
        secmenSayisi+=test20df.iat[j,1]
    tempKatilimOran = (kullananSecmenler/secmenSayisi)*100
    katilimOran=float("{0:.2f}".format(tempKatilimOran))
    return katilimOran

print("Marmaranın geçersiz oy oranı :",gecersizOranBolge('Marmara'),"Marmaranın katılım oranı :",katilimOranBolge('Marmara'))
print("Karadenizin geçersiz oy oranı :",gecersizOranBolge('Karadeniz'),"Karadenizin katılım oranı :",katilimOranBolge('Karadeniz'))
print("Ege Bölgesinin geçersiz oy oranı :",gecersizOranBolge('Ege Bölgesi'),"Ege Bölgesinin katılım oranı :",katilimOranBolge('Ege Bölgesi'))
print("Güneydoğu Anadolu geçersiz oy oranı :",gecersizOranBolge('Güneydoğu Anadolu'),"Güneydoğunun katılım oranı :",katilimOranBolge('Güneydoğu Anadolu'))
print("Doğu Anadolu geçersiz oy oranı :",gecersizOranBolge('Doğu Anadolu'),"Doğunun katılım oranı :",katilimOranBolge('Doğu Anadolu'))
print("Akdenizin geçersiz oy oranı :",gecersizOranBolge('Akdeniz Bölgesi'),"Akdenizin katılım oranı :",katilimOranBolge('Akdeniz Bölgesi'))
````
